```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
#import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
Combined_data = pd.merge(mouse_metadata, study_results, how='right')
# Display the data table for preview
Combined_data = Combined_data[['Mouse ID', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Drug Regimen', 'Sex', 'Age_months', 'Weight (g)']]
Combined_data
```

```python
# Checking the number of mice.
mice= Combined_data['Mouse ID'].value_counts()
number_of_mice=len(mice)
print("The total number of unique Mouse IDs in this study is:", number_of_mice)
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint

duplicate_mice_timepoints = Combined_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False) 

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice_ID = Combined_data.loc[duplicate_mice_timepoints,"Mouse ID"].unique()
duplicate_mice_ID 
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouse_id = Combined_data.loc[Combined_data["Mouse ID"] == "g989"]
duplicate_mouse_id
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = Combined_data[Combined_data['Mouse ID'].isin(duplicate_mice_ID)==False]
clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
updated_mice= clean_data['Mouse ID'].value_counts()
number_of_updated_mice=len(updated_mice)
print("The updated number of unique mice IDs:" , number_of_updated_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
summary_df = clean_data.groupby('Drug Regimen')

# mean, median, variance, standard deviation, and SEM of the tumor volume.
regimen_mean = summary_df["Tumor Volume (mm3)"].mean()
regimen_median = summary_df["Tumor Volume (mm3)"].median()
regimen_variance = summary_df["Tumor Volume (mm3)"].var()
regimen_std = summary_df["Tumor Volume (mm3)"].std()
regimen_sem = summary_df["Tumor Volume (mm3)"].sem()



# Assemble the resulting series into a single summary DataFrame.
summary_stats_table = pd.DataFrame({"Mean Tumor Volume": regimen_mean, 
                                    "Median Tumor Volume":regimen_median, 
                                    "Tumor Volume Variance":regimen_variance, 
                                    "Tumor Volume Std. Dev.": regimen_std, 
                                    "Tumor Volume Std. Err.": regimen_sem})

summary_stats_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_data.groupby(['Mouse ID'])['Timepoint'].max()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Combined_data2 = pd.merge(clean_data,last_timepoint,on=('Mouse ID', 'Timepoint'), how='right')
Combined_data2
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers.

# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
tumor_cap = []
tumor_ram = []
tumor_inf = []
tumor_cef = []

# Locate the rows which contain mice on each drug and get the tumor volumes

for i, row in Combined_data2.iterrows():
    if row["Drug Regimen"] == treatments[0]:
        tumor_cap.append(row['Tumor Volume (mm3)'])
    if row["Drug Regimen"] == treatments[1]:
        tumor_ram.append(row['Tumor Volume (mm3)'])
    if row["Drug Regimen"] == treatments[2]:
        tumor_inf.append(row['Tumor Volume (mm3)'])
    if row["Drug Regimen"] == treatments[3]:
        tumor_cef.append(row['Tumor Volume (mm3)'])
        
    # add subset
tumor_cap.sort()
tumor_ram.sort()
tumor_inf.sort()
tumor_cef.sort()

treatments_data=pd.DataFrame({"Capomulin":tumor_cap , 
                              "Ramicane":tumor_ram, 
                              "Infubinol":tumor_inf, 
                              "Ceftamin": tumor_cef })
treatments_data

```

```python

# Determine outliers using upper and lower bounds
quart_cap = treatments_data[treatments[0]].quantile([.25, .5, .75])
low_cap = quart_cap[.25]
med_cap = quart_cap[.5]
up_cap = quart_cap[.75]
iqr_cap = up_cap-low_cap
lowb_cap = low_cap - (1.5*iqr_cap)
upb_cap = up_cap + (1.5*iqr_cap)
max_cap = treatments_data[treatments[0]].max()
min_cap = treatments_data[treatments[0]].min()
print('Capomulin')
print('iqr:' ,iqr_cap)
print('lower bound:', lowb_cap)
print('Upper bound:' ,upb_cap)
print("Values below", lowb_cap,"could be Capomulin outliers.")
print("Values above", upb_cap,"could be Capomulin outliers.")
print("maximum capomulin:",max_cap)
print("minimum capomulin:",min_cap)

quart_ram =treatments_data[treatments[1]].quantile([.25, .5, .75])
low_ram = quart_ram[.25]
med_ram = quart_ram[.5]
up_ram = quart_ram[.75]
iqr_ram = up_ram-low_ram
lowb_ram = low_ram - (1.5*iqr_ram)
upb_ram = up_ram + (1.5*iqr_ram)
max_ram = treatments_data[treatments[1]].max()
min_ram = treatments_data[treatments[1]].min()
print('Ramicane')
print('iqr:' ,iqr_ram)
print('lower bound:', lowb_ram)
print('Upper bound:' ,upb_ram)
print("Values below", lowb_ram,"could be Ramicane outliers.")
print("Values above", upb_ram,"could be Ramicane outliers.")



quart_inf = treatments_data[treatments[2]].quantile([.25, .5, .75])
low_inf = quart_inf[.25]
med_inf = quart_inf[.5]
up_inf = quart_inf[.75]
iqr_inf = up_inf-low_inf
lowb_inf = low_inf - (1.5*iqr_inf)
upb_inf = up_inf + (1.5*iqr_inf)
max_inf = treatments_data[treatments[2]].max()
min_inf = treatments_data[treatments[2]].min()
print('Infubinol')
print('iqr:' ,iqr_inf)
print('lower bound:', lowb_inf)
print('Upper bound:' ,upb_inf)
print("Values below", lowb_inf,"could be Infubinol outliers.")
print("Values above", upb_inf,"could be Infubinol outliers.")


quart_cef = treatments_data[treatments[3]].quantile([.25, .5, .75])
low_cef = quart_cef[.25]
med_cef = quart_cef[.5]
up_cef = quart_cef[.75]
iqr_cef = up_cef-low_cef
lowb_cef = low_cef - (1.5*iqr_cef)
upb_cef = up_cef + (1.5*iqr_cef)
max_cef = treatments_data[treatments[3]].max()
min_cef = treatments_data[treatments[3]].min()
print('Ceftamin')
print('iqr:' ,iqr_cef)
print('lower bound:', lowb_cef)
print('Upper bound:' ,upb_cef)
print("Values below", lowb_cef,"could be Ceftamin outliers.")
print("Values above", upb_cef,"could be Ceftamin outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
boxplot_list = [treatments_data[treatments[0]],
                treatments_data[treatments[1]],
                treatments_data[treatments[2]],
                treatments_data[treatments[3]]
               ]
#used plt.subplots to define size
fig2,ax = plt.subplots(figsize=(9,7))
ax.set_title('Final Tumor volume in each treatment group') #set title
ax.set_xlabel('Drug Regimen') # set x label
ax.set_ylabel('Tumor Vol (mm3)') #set y lable
flierprops = dict(marker='^', markerfacecolor='b', markersize=10, markeredgecolor='black')
# used ax.boxplot function and ran the parameter boxplot_list to create the boxplot
ax.boxplot(boxplot_list, 0, flierprops=flierprops)
plt.xticks([1,2,3,4], treatments) # set labels of each boxplot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin"]

# use groupby() and mean() to create a dataframe

avg_tumor_vol= capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
avg_Mouse_weight=capomulin_df.groupby('Mouse ID')['Weight (g)'].mean()

Capomulin=pd.DataFrame({"Average tumor vol":avg_tumor_vol , 
                        "Average mouse weight":avg_Mouse_weight, 
                             })
#print(avg_tumor_vol)
#print(avg_Mouse_weight)
Capomulin
```

```python
# created scatter plot of the x axis (weight) and the y axis (tumor size) by calling from the capavg dataframe
x=avg_tumor_vol
y=avg_Mouse_weight
plt.scatter(x,y)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

#plt.savefig('scatterplot')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(x,y)
rounded_correlation = round(correlation[0], 2)
print("The correlation between mouse weight and the average tumor volume is:",rounded_correlation)

# Add the linear regression equation and line to plot
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
(slope, intercept,rvalue, pvalue, stderr)=st.linregress(Capomulin['Average mouse weight'],Capomulin['Average tumor vol'])
regress_values=Capomulin['Average mouse weight']* slope + intercept
line_eq= f"y = {round(slope, 2)} x + {round(intercept, 2)}"

plt.scatter(Capomulin['Average mouse weight'],Capomulin['Average tumor vol'],color='royalblue')
plt.plot(Capomulin['Average mouse weight'], regress_values, color='red')
plt.annotate(line_eq,(2,1), fontsize=14)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight vs Tumor Volume for Capomulin")
plt.show()
```
